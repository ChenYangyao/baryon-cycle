```python
def forward(x):
    x=np.asarray(x)
    return np.piecewise(x,
        [x<RISM0010,(x>=RISM0010)&(x<=RCGM0010),x>RCGM0010],
        [lambda x:x*50/RISM0010,
         lambda x:(x-RISM0010)*50/(RCGM0010-RISM0010)+50,
         lambda x:(x-RCGM0010)*50/(2000-RCGM0010)+100])
def inverse(y):
    y = np.asarray(y)
    return np.piecewise(y,
        [y<50,(y>=50)&(y<=100),y>100],
        [lambda y:RISM0010/50*y,
         lambda y:(RCGM0010-RISM0010)/50*(y-50)+RISM0010,
         lambda y:(2000-RCGM0010)/50*(y-100)+RCGM0010])
```
